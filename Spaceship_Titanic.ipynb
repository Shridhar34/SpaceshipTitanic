```python
train = train.dropna()
```

```python
train[['Cabin' , "CabinA" , "CabinB"]] = train['Cabin'].str.split('/',expand=True)
```

```python
train.head()
```

```python
Y = train['Transported']
X = train.drop(["Transported"] , axis = 1)
```

```python
X = X.drop(['PassengerId'] , axis = 1)
```

```python
X = X.drop(["Name"] , axis = 1)
```

```python
X['CryoSleep'] = X['CryoSleep'].replace({True:1 , False:0})
```

```python
X.head()
```

```python
df_encoded = pd.get_dummies(data = X , columns = ["HomePlanet"])
X.drop(columns=['HomePlanet'], inplace=True)
X = pd.concat([X, df_encoded], axis=1)
```

```python
X.head()
```

```python
df_encoded = pd.get_dummies(data = X , columns = ["Cabin"])
X.drop(columns=['Cabin'], inplace=True)
X = pd.concat([X, df_encoded], axis=1)


df_encoded = pd.get_dummies(data = X , columns = ["CabinB"])
X.drop(columns=['CabinB'], inplace=True)
X = pd.concat([X, df_encoded], axis=1)
```

```python
df_encoded = pd.get_dummies(data = X , columns = ["Destination"])
X.drop(columns=['Destination'], inplace=True)
X = pd.concat([X, df_encoded], axis=1)
```

```python
X['VIP'] = X['VIP'].replace({True:1 , False:0})
```

```python
Y = Y.replace({True:1 , False:0})
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
rf_classifier1 = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier1.fit(X, Y)
```

```python
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
log_reg = LogisticRegression()

# Train the model
log_reg.fit(X, Y)
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

dt = DecisionTreeClassifier(random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X, Y, test_size=0.2, random_state=42)
dt.fit(X_train, y_train)
# Initialize the Decision Tree Classifier
dt = DecisionTreeClassifier(random_state=42)

# Train the model without pruning
dt.fit(X_train, y_train)

# Predict the results on validation set
y_pred = dt.predict(X_val)

# Evaluate the model without pruning
accuracy = accuracy_score(y_val, y_pred)
conf_matrix = confusion_matrix(y_val, y_pred)
class_report = classification_report(y_val, y_pred)

print(f"Accuracy before pruning: {accuracy:.4f}")
print(f"Confusion Matrix before pruning:\n{conf_matrix}")
print(f"Classification Report before pruning:\n{class_report}")

# Applying Cost Complexity Pruning
path = dt.cost_complexity_pruning_path(X_train, y_train)  # Getting the alpha values for pruning
ccp_alphas, impurities = path.ccp_alphas, path.impurities

# Loop through different alpha values and train the models
clfs = []
for ccp_alpha in ccp_alphas:
    clf = DecisionTreeClassifier(random_state=42, ccp_alpha=ccp_alpha)
    clf.fit(X_train, y_train)
    clfs.append(clf)

# Evaluate the models on the validation set and choose the best one
best_clf = None
best_accuracy = 0
for clf in clfs:
    y_pred = clf.predict(X_val)
    accuracy = accuracy_score(y_val, y_pred)
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_clf_decision_tree = clf

# Evaluate the best pruned model
y_pred = best_clf_decision_tree.predict(X_val)

# Evaluate the pruned model
print(f"Accuracy after pruning: {best_accuracy:.4f}")
print(f"Confusion Matrix after pruning:\n{confusion_matrix(y_val, y_pred)}")
print(f"Classification Report after pruning:\n{classification_report(y_val, y_pred)}")

# Plotting the best pruned tree
#plt.figure(figsize=(15,10))
#plot_tree(y_pred = best_clf_decision_tree.predict(X_val)
#, filled=True, feature_names=X.columns, class_names=['Not Transported', 'Transported'])
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}
rf_classifier = RandomForestClassifier(random_state=42)
random_search = RandomizedSearchCV(estimator=rf_classifier, param_distributions=param_grid,
                                   n_iter=10, scoring='accuracy', cv=5, random_state=42, n_jobs=-1)
random_search.fit(X, Y)
best_params = random_search.best_params_
print("Best Hyperparameters:", best_params)
best_rf_model = random_search.best_estimator_
```
